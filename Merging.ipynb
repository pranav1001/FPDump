```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
hrv=pd.read_csv(r'D:\Dataset\3 - Feature dataset\per sensor\D - Physiology features (HR_HRV_SCL - final).csv')
facial_features=pd.read_csv(r'facial features.csv')
```

```python
facial_features.drop_duplicates(inplace=True)
facial_features.replace(999,np.nan,inplace=True)
facial_features.dropna(inplace=True)
hrv.replace(999,np.nan,inplace=True)
```

```python
features=pd.merge(facial_features,hrv, how='inner', on=['timestamp','PP','C'])
```

```python
features_na=features[features['HR'].isna()]
```

```python
features=features[~features['HR'].isna()]
```

```python
x=features[['Squality', 'Sneutral', 'Shappy', 'Ssad',
       'Sangry', 'Ssurprised', 'Sscared', 'Sdisgusted', 'Svalence',
       'SyHeadOrientation', 'SxHeadOrientation', 'SzHeadOrientation',
       'SmouthOpen', 'SleftEyeClosed', 'SrightEyeClosed',
       'SleftEyebrowLowered', 'SleftEyebrowRaised', 'SrightEyebrowLowered',
       'SrightEyebrowRaised', 'SgazeDirectionForward', 'SgazeDirectionLeft',
       'SgazeDirectionRight', 'SAu01_InnerBrowRaiser', 'SAu02_OuterBrowRaiser',
       'SAu04_BrowLowerer', 'SAu05_UpperLidRaiser', 'SAu06_CheekRaiser',
       'SAu07_LidTightener', 'SAu09_NoseWrinkler', 'SAu10_UpperLipRaiser',
       'SAu12_LipCornerPuller', 'SAu14_Dimpler', 'SAu15_LipCornerDepressor',
       'SAu17_ChinRaiser', 'SAu20_LipStretcher', 'SAu23_LipTightener',
       'SAu24_LipPressor', 'SAu25_LipsPart', 'SAu26_JawDrop',
       'SAu27_MouthStretch', 'SAu43_EyesClosed']].to_numpy()
y=features['HR'].to_numpy()
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler((-1,1))
X_norm = min_max_scaler.fit_transform(x)
y_norm=min_max_scaler.fit_transform(y.reshape(-1,1))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_norm, y_norm, test_size=0.25, random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(X_train, y_train)
reg.score(X_test,y_test)
```
